```python
# import packages
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib

# Load user_reviews.csv
reviews_df = pd.read_csv("march/boxplot_28_march.csv")
#print(reviews_df.head()) 

sns.set_style('ticks')
fig, ax = plt.subplots()
fig.set_size_inches(15, 8)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.setp(ax.get_xticklabels(), fontsize=16)


font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 16}

matplotlib.rc('font', **font)

#User review sentiment polarity for paid vs. free apps
ax = sns.boxplot(x = reviews_df['Weeks'], y = reviews_df['Type'])
ax.set_title('Average Duration in weeks for each Project Stage', fontsize=20)

plt.savefig('march.png', bbox_inches="tight")
```
